```python
s=8
k=np.array([[1,5,8,10,12,14,16,18,20,25,30,35,40,45,50]])
# pick a certain sparsity level for s, and calculate the standard deviation
#----------------------------------
dense_train_acc=[]
dense_test_acc=[]
sparse_train_acc=[]
sparse_test_acc=[]
for width in k:
    dense_history=TrainHistory.load('result/pretrain_{}.history'.format(s,width))
    dense_test_acc.append([width,1-dense_history.test_acc[-1]])
    dense_train_acc.append([width,1-dense_history.train_acc[-1]]) #calculate the error
    if s<=width:
        sparse_history=TrainHistory.load('result/sparse_{}_s={}.history'.format(width,s))
        sparse_test_acc.append([width,1-sparse_history.test_acc[-1]])
        sparse_train_acc.append([width,1-sparse_history.train_acc[-1]])

        
dense_test_acc=np.array(dense_test_acc)
dense_train_acc=np.array(dense_train_acc)
sparse_test_acc=np.array(sparse_test_acc)
sparse_train_acc=np.array(sparse_train_acc)
```
